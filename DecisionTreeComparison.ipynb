```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
#changing categorical to numerical value
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
#display input variables 
independent
```

```python
dependent=dataset[["Profit"]]
```

```python
#output variable display
dependent
```

```python
#split test and training data

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#model creation step - first two step is procedure after which model is created
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson', splitter='random')
regressor=regressor.fit(X_train,y_train)
```

```python
#matplotlib helps plotting the graph
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred) 
```
